```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import gradio as gr

# Load model and tokenizer
model_name = "EleutherAI/gpt-neo-125M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Chat function
def career_advisor(question):
    input_ids = tokenizer.encode(question, return_tensors="pt")
    response = model.generate(
        input_ids, max_length=100, num_return_sequences=1, do_sample=True
    )
    output = tokenizer.decode(response[0], skip_special_tokens=True)
    return output.strip()

# Gradio interface
gr.Interface(
    fn=career_advisor,
    inputs="text",
    outputs="text",
    title="Career Advisor Chatbot",
    description="Ask career-related questions about jobs, skills, or resumes!"
).launch()
```

```python
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments

# Convert to DataFrame
df = pd.DataFrame(data)

# Combine question and answer into a single 'text' field
df['text'] = df.apply(lambda row: f"Question: {row['question']}\nAnswer: {row['answer']}", axis=1)

# Convert to Dataset
dataset = Dataset.from_pandas(df[['text']])  # Only keep the 'text' column

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define a pad_token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token  # Use eos_token as the pad_token

def tokenize_function(examples):
    # Tokenize the input and create labels
    tokenized = tokenizer(examples["text"], truncation=True, padding="max_length", max_length=512)
    tokenized["labels"] = tokenized["input_ids"].copy()  # Set labels to match input_ids
    return tokenized

# Tokenize the dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)



tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split into train/test datasets
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.1)

# Load pre-trained model
model = AutoModelForCausalLM.from_pretrained(model_name)

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    eval_steps=100,
    logging_dir="./logs",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    save_steps=100,
    save_total_limit=2,
    learning_rate=5e-5,
    warmup_steps=50,
    report_to=[],  # Disable all logging integrations, including wandb
)


# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"]
)

# Fine-tune the model
trainer.train()

# Save the fine-tuned model
model.save_pretrained("./career_advisor_model")
tokenizer.save_pretrained("./career_advisor_model")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import gradio as gr

# Load fine-tuned model and tokenizer
model_name = "./career_advisor_model"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Define the chatbot function
def career_advisor(question, temperature=0.7, max_length=100):
    """
    Generate a response from the fine-tuned model based on user input.
    Args:
        question (str): User's input question.
        temperature (float): Controls randomness in the model's output.
        max_length (int): Maximum length of the response.
    Returns:
        str: Model's response.
    """
    input_ids = tokenizer.encode(question, return_tensors="pt")
    response = model.generate(
        input_ids,
        max_length=max_length,
        num_return_sequences=1,
        do_sample=True,
        temperature=temperature
    )
    output = tokenizer.decode(response[0], skip_special_tokens=True)
    return output.strip()

# Gradio interface
interface = gr.Interface(
    fn=career_advisor,
    inputs=[
        gr.Textbox(
            lines=3,
            label="Ask a Career-Related Question",
            placeholder="e.g., What skills do I need to become a data scientist?",
        ),
        gr.Slider(0.1, 1.0, value=0.7, label="Response Creativity (Temperature)"),
        gr.Slider(50, 300, value=100, label="Max Response Length"),
    ],
    outputs=gr.Textbox(
        lines=5, label="Response"
    ),
    title="Career Advisor Chatbot",
    description=(
        "Ask career-related questions about jobs, resumes, or skills! "
        "This chatbot is fine-tuned to provide tailored advice."
    ),
    examples=[
        ["What jobs are good for data science skills?", 0.7, 100],
        ["How can I improve my resume?", 0.6, 120],
        ["What are the top skills for a data analyst?", 0.8, 150],
    ],
    theme="compact",
)


interface.launch()
```
